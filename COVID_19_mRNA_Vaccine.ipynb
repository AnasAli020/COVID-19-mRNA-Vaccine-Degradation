##  Feature Engineering & Encoding

### Feature Extraction

```python
df_traiin.head()
```

```python
df_traiin[['predicted_loop_type_a','predicted_loop_type_b','sequence_a','sequence_b','structure_a','structure_b']]=0

for i in range(len(df_traiin)):

  df_traiin['predicted_loop_type_a'][i]=f"{df_traiin['a1_predicted_loop_type'][i]}{df_traiin['a2_predicted_loop_type'][i]}{df_traiin['a3_predicted_loop_type'][i]}{df_traiin['a4_predicted_loop_type'][i]}{df_traiin['a5_predicted_loop_type'][i]}"
  df_traiin['predicted_loop_type_b'][i]=f"{df_traiin['b1_predicted_loop_type'][i]}{df_traiin['b2_predicted_loop_type'][i]}{df_traiin['b3_predicted_loop_type'][i]}{df_traiin['b4_predicted_loop_type'][i]}{df_traiin['b5_predicted_loop_type'][i]}"
  df_traiin['sequence_a'][i]=f"{df_traiin['a1_sequence'][i]}{df_traiin['a2_sequence'][i]}{df_traiin['a3_sequence'][i]}{df_traiin['a4_sequence'][i]}{df_traiin['a5_sequence'][i]}"
  df_traiin['sequence_b'][i]=f"{df_traiin['b1_sequence'][i]}{df_traiin['b2_sequence'][i]}{df_traiin['b3_sequence'][i]}{df_traiin['b4_sequence'][i]}{df_traiin['b5_sequence'][i]}"
  df_traiin['structure_a'][i]=f"{df_traiin['a1_structure'][i]}{df_traiin['a2_structure'][i]}{df_traiin['a3_structure'][i]}{df_traiin['a4_structure'][i]}{df_traiin['a5_structure'][i]}"
  df_traiin['structure_b'][i]=f"{df_traiin['b1_structure'][i]}{df_traiin['b2_structure'][i]}{df_traiin['b3_structure'][i]}{df_traiin['b4_structure'][i]}{df_traiin['b5_structure'][i]}"

df_traiin.drop(['a1_predicted_loop_type','a2_predicted_loop_type','a3_predicted_loop_type','a4_predicted_loop_type','a5_predicted_loop_type',
               'b1_predicted_loop_type','b2_predicted_loop_type','b3_predicted_loop_type','b4_predicted_loop_type','b5_predicted_loop_type',
               'a1_sequence','a2_sequence','a3_sequence','a4_sequence','a5_sequence',
               'b1_sequence','b2_sequence','b3_sequence','b4_sequence','b5_sequence',
               'a1_structure','a2_structure','a3_structure','a4_structure','a5_structure',
               'b1_structure','b2_structure','b3_structure','b4_structure','b5_structure'],inplace=True,axis=1)
sequence_b_counts=[]
for i in df_traiin['sequence_b'] :
  sequence_b_counts.append(len(i))
sequence_a_counts=[]
for i in df_traiin['sequence_a'] :
  sequence_a_counts.append(len(i))

predicted_loop_type_a_counts=[]
for i in df_traiin['predicted_loop_type_a'] :
  predicted_loop_type_a_counts.append(len(i))
predicted_loop_type_b_counts=[]
for i in df_traiin['predicted_loop_type_b'] :
  predicted_loop_type_b_counts.append(len(i))

structure_a_counts=[]
for i in df_traiin['structure_a'] :
  structure_a_counts.append(len(i))
structure_b_counts=[]
for i in df_traiin['structure_b'] :
  structure_b_counts.append(len(i))

df_traiin['sequence_a_counts']=sequence_a_counts
df_traiin['sequence_b_counts']=sequence_b_counts
df_traiin['predicted_loop_type_a_counts']=predicted_loop_type_a_counts
df_traiin['predicted_loop_type_b_counts']=predicted_loop_type_b_counts
df_traiin['structure_a_counts']=structure_a_counts
df_traiin['structure_b_counts']=structure_b_counts
df_traiin.drop(['predicted_loop_type_a',	'predicted_loop_type_b'	,'sequence_a',	'sequence_b',	'structure_a'	,'structure_b'],inplace=True,axis=1)
```

### Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
cat_data
OE = cat_data.columns
dic = {'A':0, 'G': 1, 'U': 2, 'C': 3,
       '.': 1, '(': 2, ')':3,
       'B':0, 'E': 1, 'H': 2, 'I': 3, 'M': 4, 'S': 5, 'X': 6,'-1':0}
df_train[OE] = df_train[OE].replace(dic)
test[OE] = test[OE].replace(dic)
```

## **Target1**

```python
# Split Data
x_train,x_test,y_train, y_test = train_test_split(df_train,target1, test_size=0.25, random_state=94)
```

### **modeling**

```python
model=GradientBoostingRegressor(learning_rate=0.302, n_estimators=500,criterion='squared_error',min_samples_leaf=5,max_depth=4)
GB_model =model.fit(x_train,y_train)
```

```python
pred= GB_model.predict(x_test) # random_state= 94 - lr= 0.302 -  ns= 500 -  squared_error  - min_samples_leaf= 5 - max_depth=4
np.sqrt(mean_squared_error(pred, y_test))
```

```python
predection1=GB_model.predict(test)
```
